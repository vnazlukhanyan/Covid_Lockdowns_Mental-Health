```python
death = [(u'0 - 9 Years', 94), (u'40 - 49 Years', 6872), 
        (u'80+ Years', 112347), (u'50 - 59 Years', 17808), 
        (u'60 - 69 Years', 38550), (u'70 - 79 Years', 59352), (u'10 - 19 Years', 207), 
        (u'30 - 39 Years', 2900), (u'20 - 29 Years', 1132)]

death_sorted = sorted(death, key=lambda r:r[0])
```

```python
total = [(u'0 - 9 Years', 527700), (u'40 - 49 Years', 2013231),
         (u'80+ Years', 579595), (u'50 - 59 Years', 1945338),
         (u'60 - 69 Years', 1372151), (u'70 - 79 Years', 762250), (u'10 - 19 Years', 1367283), 
         (u'30 - 39 Years', 2198860), (u'20 - 29 Years', 2567606)]

total_sorted = sorted(total, key=lambda r:r[0])
```

```python
count = {'Age_Group': [],
        'Death_Count': [],
        'Total_Count': []}
for row in death_sorted:
    count['Age_Group'].append(row[0])
    count['Death_Count'].append(row[1])
for row in total_sorted:
    count['Total_Count'].append(row[1])

count
```

```python
df_count = pd.DataFrame(count)
df_count['death_rate'] = df_count['Death_Count'] / df_count['Total_Count']
df_count
```

```python
fig, ax = plt.subplots(figsize=(12,6))
y_pos = np.arange(9)

for i in y_pos:
    if i == 8:
        ax.plot([0, max(df_count['Total_Count'])+200], [i, i],':', color = 'lightgrey')
        ax.plot([0, df_count['Total_Count'][i]], [i, i] ,color = 'grey')
        ax.text(-450000,i,df_count['Age_Group'][i], verticalalignment='center', horizontalalignment='left')
        ax.text(300000,i+0.1, f"{df_count['death_rate'][i]:.3%}", color='red', size=12)
        ax.text(df_count['Total_Count'][i]+50000, i-0.1, df_count['Total_Count'][i])
        
    else:
        ax.plot([0, max(df_count['Total_Count'])+200], [i, i],':', color = 'lightgrey')
        ax.plot([0, df_count['Total_Count'][i]], [i, i] ,color = 'grey')
        ax.text(-450000,i,df_count['Age_Group'][i], verticalalignment='center',horizontalalignment='left')
        ax.text(300000,i+0.1, f"{df_count['death_rate'][i]:.3%}")
        ax.text(df_count['Total_Count'][i]+50000, i-0.1, df_count['Total_Count'][i])


    ax.hlines(y=i, xmin=-5, xmax=0, color='lightgrey',lw=0.4)
    
ax.vlines(x=0, ymin=0, ymax=8, color='dimgray', lw=0.1)
ax.text(0, 8.8, 'Death Rate -- Confirmed Cases',fontsize=14)
ax.axis('off')
fig.show()
```
